```python
data1 = data.drop(columns = ['ID', 'Year_Birth', 'Education', 'Marital_Status', 'Kidhome',
       'Teenhome', 'Dt_Customer', 'Recency', 'MntWines',
       'MntMeatProducts', 'MntFishProducts', 'MntSweetProducts',
       'MntGoldProds', 'NumDealsPurchases', 'NumWebPurchases',
       'NumCatalogPurchases', 'NumStorePurchases', 'NumWebVisitsMonth',
       'AcceptedCmp3', 'AcceptedCmp4', 'AcceptedCmp5', 'AcceptedCmp1',
       'AcceptedCmp2', 'Complain', 'Z_CostContact', 'Z_Revenue', 'Response'])
```

```python
data1.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
min_max_scaler = MinMaxScaler()
min_max_scaler.fit(data1)
data1 = pd.DataFrame(min_max_scaler.transform(data1))
data1.columns = ['Income','MntFruits']
data1.head()
```

```python
data1.info()
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=3)
```

```python
kmeans.fit(data1)
```

```python
kmeans.cluster_centers_
```

```python
y = kmeans.labels_
print(y)
```

```python
data1['Cluster'] = y
```

```python
data1.head(100)
```

```python
y.shape
```

```python
sns.countplot(x = kmeans.labels_, palette = 'Oranges')
```

```python
plt.figure(figsize=(10,10))
sns.scatterplot(data1['Income'].values,data['MntFruits'].values, c=kmeans.labels_)
centers = np.array(kmeans.cluster_centers_)
plt.scatter(centers[:,0],centers[:,1],marker = 'x',color ='k')
plt.show()
```

```python
sns.scatterplot(data1['Income'], data1['MntFruits'], hue = data1['Cluster'])
```

```python
k_rng = range(1,10)
sse = []
for k in k_rng:
    km = KMeans(n_clusters=k)
    km.fit(data1[['Income','MntFruits']])
    sse.append(km.inertia_)
```

```python
sse
```

```python
plt.xlabel('K')
plt.ylabel('Sum of squared error')
plt.plot(k_rng, sse)
```
